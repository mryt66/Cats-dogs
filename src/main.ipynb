```python
import numpy as np
import pandas as pd
import random
import csv
import os
from statistics import mean
from csv_creator import DogCat

class LoadData:
    def __init__(self, data):
        self.data = data
        
    def shuffle(self):
        for i in range(len(self.data) - 1, 0, -1):
            j = random.randint(0, i)
            self.data[i], self.data[j] = self.data[j], self.data[i]
        return self.data
    
    def split(self):
        split = int(len(self.data)*0.5)
        listTrain = self.data[0:split]
        listVal = self.data[split: len(self.data)]
        return listTrain, listVal
    

class KNN:
    properly_estimated = 0
       
    def __init__(self, train, val, k):
        self.train = train
        self.val = val
        self.k = k
        
    @staticmethod
    def euclidean_dist(x1, x2):
        x1 = np.array(x1[:-1])
        x2 = np.array(x2[:-1])
        return np.sqrt(np.sum((x1 - x2)**2))
    
    def classify(self):
        tmp1 = []
        for i in self.val:
            tmp2 = [[self.euclidean_dist(i,j),j[4]] for j in self.train]
            tmp2.sort()
            print(tmp2)
            tmp1.append(tmp2)
        
        for i in range(len(tmp1)):
            cats = 0
            dogs = 0
            for j in range(0, self.k):
                if tmp1[i][j][1]=='cats':
                    cats+=1
                else:
                    dogs+=1
                    
            if cats>dogs and self.val[i][4]=='cats':
                self.properly_estimated+=1
            elif dogs>cats and self.val[i][4]=='dogs':
                self.properly_estimated+=1
                
        print(f'Skuteczność klasyfikatora wynosi około {round((self.properly_estimated/len(tmp1)*100),2)}%')


def LoadCsv(file, type):
    with open(file, newline='') as file:
        reader = csv.reader(file, quoting=csv.QUOTE_NONE)
        for i in reader:
            tmp=[float(j) for j in i]
            num = 0
            for j in tmp:
                num+=float(j)
            if len(tmp)+1>5:
                data.append([mean(tmp), min(tmp), max(tmp), len(tmp), type])
    return data



cats_csv='Data_cats.csv'
dogs_csv='Data_dogs.csv'

if os.path.getsize(cats_csv)==0 and os.path.getsize(dogs_csv)==0:
    for i in range(26):
        thing = DogCat('cat', i, 'Data_cats.csv')
        thing.read_wave()
        thing.create_data()

        thing = DogCat('dog_barking', i, 'Data_dogs.csv')
        thing.read_wave()
        thing.create_data()
        

data=[]
LoadCsv(cats_csv,'cats')
LoadCsv(dogs_csv,'dogs')

x = LoadData(data)
x.shuffle()
train, val = x.split()

k = 5
y = KNN(train,val,k)
y.classify()

```
